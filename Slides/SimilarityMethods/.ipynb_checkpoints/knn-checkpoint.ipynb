### K Nearest Neighbors

In these slides we explore a simple but powerful learning algorithm called $K$ Nearest Neighbors (KNN).

KNN is an [instance based](https://en.wikipedia.org/wiki/Instance-based_learning) method that can be used for both classification and regression.





```python
import numpy as np
from scipy.stats import mode


def euclidean(X,x):
    return np.linalg.norm(X-x,axis=1)



def KNN(X,x,k=3,distance = euclidean):
    """Return the row indexes of the k nearest neighbors of x
       in the matrix X"""
    distances = distance(X,x)
    neighbors = []
    for i in range(k):
        nearest = distances.argmin()
        neighbors.append(nearest)
        distances[nearest] = distances.max()
    return neighbors

def KNN_predict_all(X_known,y_known,X_predict,k=3,distance=euclidean,regression= False):
    y_hat = np.empty(X_predict.shape[0])
    for i,x in enumerate(X_predict):
        knn = KNN(X_known,x,k,distance)
        if regression:
            y_hat[i] = np.mean(y_known[knn])
        else:
            y_hat[i] = mode(y_known[knn])[0][0]
    return y_hat

```

```python
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs

X,y = make_blobs(centers=2,random_state=42,cluster_std=4)

y = y*2-1


import matplotlib.pyplot as plt


Xa = X[y==1]
Xb = X[y==-1]
plt.scatter(Xa[:,0],Xa[:,1])
plt.scatter(Xb[:,0],Xb[:,1])

special = np.array([3,-1])
plt.scatter([special[0]],[special[1]],marker="*",s=80,label="point to classify")
plt.legend()
plt.title("Classification training data")
plt.show()
```

```python
neighbors = KNN(X,special)
X[neighbors]
```

```python

Xa = X[y==1]
Xb = X[y==-1]
plt.scatter(Xa[:,0],Xa[:,1])
plt.scatter(Xb[:,0],Xb[:,1])

plt.scatter([special[0]],[special[1]],marker="*",s=80,label="point to classify")
plt.legend()

plt.scatter(X[neighbors][:,0],X[neighbors][:,1],s=120,alpha=0.2,label="neighbors")
plt.title("Classification training data")
plt.show()
```

```python


k = 1


def plot_boundary(X,y,k,num=100):
    xc = np.linspace(X[:,1].min(),X[:,1].max(),num)
    yc = np.linspace(X[:,0].min(),X[:,0].max(),num)
    cart_prod = np.transpose([np.tile(xc, len(yc)), np.repeat(yc, len(xc))])
    Xc = np.c_[cart_prod[:,1],cart_prod[:,0]]

    g = KNN_predict_all(X,y,Xc,k=k)
    #nn_predict_all(Xc,w,regression)
    says_yeah = g == 1
    says_no = ~says_yeah

    plt.title("Decision boundary of KNN, K={}".format(k))

    plt.scatter(Xc[says_yeah][:,0],Xc[says_yeah][:,1],alpha=0.05,color='g')
    plt.scatter(Xc[says_no][:,0],Xc[says_no][:,1],alpha=0.05,color='yellow')
    plt.scatter(X[y==1][:,0],X[y==1][:,1],label="1",alpha=0.43,c='b')
    plt.scatter(X[y!=1][:,0],X[y!=1][:,1],label="-1",alpha=0.43,c='red')
    plt.legend()
    plt.show()

    g = KNN_predict_all(X,y,X,k=k)
    print("Training error = ",np.mean(g != y))


for k in [1,3,8,15,25]:    
    plot_boundary(X,y,k)
```

```python
X,y = make_moons(noise=0.1)
for k in [1,3,8,15,25]:    
    plot_boundary(X,y,k)
```

### Real data

We now try out our KNN model on real data.

We will use the famous [iris](https://en.wikipedia.org/wiki/Iris_flower_data_set) dataset. 

KNN naturally does [multiclass classification](https://en.wikipedia.org/wiki/Multiclass_classification), so we will do our first multiclass classification exercise.



```python
from sklearn import datasets

# import some data to play with
iris = datasets.load_iris()
X,y = iris['data'],iris['target']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=None)
plt.hist(y_test)
plt.show()
```

```python
y_hat = KNN_predict_all(X_train,y_train,X_test)
y_hat
```

```python
y_test
```

```python
np.mean(y_hat != y_test)
```

```python
X_iris2d = X[:,[1,3]]
Xa = X_iris2d[y==1]
Xb = X_iris2d[y==0]
Xc = X_iris2d[y==2]
plt.figure(figsize=(8,8))
plt.scatter(Xa[:,0],Xa[:,1],label="versicolor")
plt.scatter(Xb[:,0],Xb[:,1],label="setosa")
plt.scatter(Xc[:,0],Xc[:,1],label="virginica")

plt.title("Classification training data")
plt.legend()
plt.show()
```

```python

def plot_multilabel(X,y,k,num=100):
    xc = np.linspace(X[:,1].min(),X[:,1].max(),num)
    yc = np.linspace(X[:,0].min(),X[:,0].max(),num)
    cart_prod = np.transpose([np.tile(xc, len(yc)), np.repeat(yc, len(xc))])
    Xc = np.c_[cart_prod[:,1],cart_prod[:,0]]

    g = KNN_predict_all(X,y,Xc,k=k)
    for gg in set(g):
        says_gg = g == gg
        plt.scatter(Xc[says_gg][:,0],Xc[says_gg][:,1],alpha=0.05)
        
        
    plt.title("Decision boundary of KNN, K={}".format(k))


    for yy in set(y):
        plt.scatter(X[y==yy][:,0],X[y==yy][:,1],label="{}".format(yy),alpha=0.75)
    
    plt.legend()
    plt.show()

    g = KNN_predict_all(X,y,X,k=k)
    print("Training error = ",np.mean(g != y))

for k in [1,3,6,10,15,25]:    
    plot_multilabel(X_iris2d,y,k)
```

###  Regression

KNN can be used for regression as easily as classification.

Rather than take the mode of the neigboring labels, we simply take their mean.


```python
from sklearn.datasets import make_regression

xx = np.linspace(0,5)
X = np.random.choice(xx,100)
y = np.sin(X)+ np.random.randn(len(X))/4
plt.scatter(X,y,alpha=0.6)
plt.show()
```

```python
# KNN_predict_all(X_known,y_known,X_predict,k=3,distance=euclidean,regression= False):

for k in [1,3,6,10,15,20]:
    yhat = KNN_predict_all(X.reshape(-1,1),y,xx,regression=True,k=k)
    plt.scatter(X,y,alpha=0.6)
    plt.plot(xx,yhat,c='r')
    plt.title("KNN regression with k = {}".format(k))
    plt.show()
```

### A real dataset

Here we apply KNN regression to the concrete dataset.


```python
import pandas as pd

df = pd.read_excel("Concrete_Data.xls")
df.head()
```

```python

X = df.values[:,:-1]
y = df.values[:,-1]

X_train,X_test,y_train,y_test = train_test_split(X,y)

from sklearn.metrics import r2_score

for k in [1,3,6,10,15,20]:

    yhat = KNN_predict_all(X_train,y_train,X_test,k=k)


    score_test = r2_score(yhat,y_test)
    print("Performance of k={} is {}".format(k,score_test))
    

```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(weights="distance",p=2)

knn.fit(X_train,y_train)

knn.score(X_test,y_test)
```
