### Stochastic Gradient Descent (SGD)

Consider a typical error function, such as ordinary least squares:

$$E_{in}(\bar{w}) = \frac{1}{N}\sum_{n=1}^N (\bar{w}^T\bar{x}_n - y_n)^2$$

We want to notice two things.  

**One thing** is that this is just the average of the pointwise error function

$${\bf{e}}(\bar{x}_n,y_n) = (\bar{w}^T\bar{x}_n - y_n)^2$$

**The second thing** is that the gradient of $E_{in}$ depends on every point in the dataset.

For instance, the derivative of $E_{in}$ with respect to $w_k$ is

$$\frac{\partial}{\partial w_k} E_{in} = \frac{2}{N}\sum_{n=1}^N (\bar{w}^T\bar{x}_n - y_n)x_k$$

To compute this **for a single $\bar{w}$** we need to know $\sum_{n=1}^N (\bar{w}^T\bar{x}_n - y_n)x_k$, which depends on every data point.

If $N=10^6$ or so this will take awhile, especially if $d$ is also large.

And then for gradient descent we might need to do this for thousands of $\bar{w}$.

This is easily getting into the billions of computational steps.

The takeaway:  **Gradient descent can be too slow**

### Stochastic approach

Stochastic is a fancy word for "random".

The idea in SGD is to use $\nabla {\bf e}$ in place of $\nabla E_{in}$.

Note that when 

$${\bf{e}}(\bar{x}_n,y_n) = (\bar{w}^T\bar{x}_n - y_n)^2$$

then

$$\nabla_{\bar{w}}{\bf{e}}(\bar{x}_n,y_n) = 2(\bar{w}^T\bar{x}_n - y_n)\bar{x}_n$$

depends only on $\bar{x}_n$, a single datapoint.

Observe that if $\bar{x}_n$ is selected from $X$ uniformly at random, then the **expected value** of $\nabla {\bf e}$ is $\nabla E_{in}$.

$$\mathbb{E}[\nabla_{\bar{w}}{\bf{e}}(\bar{x}_n,y_n)] = \frac{1}{N}\sum_{n=1}^N 2(\bar{w}^T\bar{x}_n - y_n)\bar{x}_n.$$

But the right hand side is exactly $\nabla E_{in}$.

#### SGD

This lets us make the following change to gradient descent.

Rather than the update rule

$$\bar{w} = \bar{w} - \eta \nabla E_{in}{\bar{w}}$$

we do

$$\bar{w} = \bar{w} - \eta \nabla {\bf e}(\bar{x}_n,y_n){\bar{w}}$$

This is faster by a factor of $N$. 

And if $\bar{x}_n$ is randomly chosen then this *on average* will go in the same direction as regular gradient descent.

#### Pseudocode for SGD

The pseudocode for SGD is as follows

1. Randomly initialize $\bar{w}$
2. Shuffle the rows of $X$
3. For each row $\bar{x}$ in $X$ perform the update 
$$\bar{w} = \bar{w} - \eta \nabla_{\bar{w}} {\bf e}(\bar{x}_n,y_n)$$

4. Perform more iterations through $X$ if necessary

#### Behavior

The effect of SGD is for the "ball" to "roll downhill" just as in batch gradient descent.

But now the ball only goes downhill "on average".

Usually the extra bouncing requires us to make $\eta$ an order of magnitude smaller to achieve convergence.


### Example

Below we will look at some examples of applying SGD to regression and classification problems.

Because we want to visualize a two dimensional weight space, we need one dimensional data.

```python
import mystuff as ms
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(100)

ynum=20  ## Number of positive examples
nnum=15  ## Number of negative examples

yes = np.random.randn(ynum)*3-2  ## positive examples
no = np.random.randn(nnum)+3     ## negative examples

yyes = np.ones(ynum)
yno = np.ones(nnum)*-1

preX = np.hstack((yes,no))  ## X but no bias yet
y = np.hstack((yyes,yno))   ## targets y

X = np.c_[np.ones(ynum+nnum),preX]  ## adding bias

w = np.array([4,-3])
fig = plt.figure(figsize=(9,8))

plt.scatter(X[:ynum,1],np.zeros(ynum),c='b',alpha=0.3,s=60,label="positive example")
plt.scatter(X[ynum:,1],np.zeros(nnum),c='r',alpha=0.3,s=60,label="negative example")

plt.plot(X[:,1],X.dot(w))
plt.title("Some 1D data and a decision boundary")
plt.legend()
plt.show()
```

```python
def E_in(X,w,y):
    mc = (np.sign(X.dot(w)) != y)
    return np.mean(mc)
E_in(X,w,y)
```

### Classification error

Above we compute the usual in-sample error for a classification problem.

$$E_{in} = \frac{1}{N}\sum_{n=1}^N [\![sign(\bar{w}^T\bar{x}_n) \neq y_n ]\!]$$

Below we visualize the plot of $E_{in}$ using a contour plot.

Notice that the derivative of $sign(s)$ is always either zero or undefined.

The plot of this $E_{in}$ is discontinuous.

It is like a pie where each pie piece is flat, but also each piece of the pie has a different height. 

This makes it hard to minimize $E_{in}$ using gradient descent (or any other method).

The ball will not roll.



```python

xx = np.linspace(-10,10,100)
yy = np.linspace(-10,10,100)

XX,YY = np.meshgrid(xx,yy)
W = np.c_[XX.ravel(),YY.ravel()]
errs = np.sign(X.dot(W.T)).T != y
E = np.mean(errs,axis=1)
plt.contourf(XX,YY,E.reshape(100,100),levels=300)
plt.title(r"Contour plot of $z=E_{in}(\bar{w})$ (classification)")
plt.xlabel(r"$w_0$")
plt.ylabel(r"$w_1$")
plt.colorbar()
plt.show()
```

### Regression for classification

Here is a trick we can use to get a $E_{in}$ that is easy to optimize.

We consider $\mathcal{Y} = \{\pm 1\}$ not as classes, but as outputs of the target function.

We will use regression to find a line fitting the data. 

The error that we minimize to find this $\bar{w}$ is the OLS error from regression (see first slide).

**But** having found this $\bar{w}$ we then use it for classification using the rule

$$\hat{y} = sign(\bar{w}^T\bar{x})$$

Put another way, we **threshold** $\bar{w}^T\bar{x}$ with the threshold $\theta=0$.

$$h_\bar{w}(\bar{x}) = \begin{cases} +1 \,if\, \bar{w}^T\bar{x} > \theta \\ 
                                    -1 \,if\, \bar{w}^T\bar{x} \leq \theta\end{cases}$$
                                    
We plot this situation below.


```python
import mystuff as ms
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(100)

ynum=20  ## Number of positive examples
nnum=15  ## Number of negative examples

yes = np.random.randn(ynum)*3-2  ## positive examples
no = np.random.randn(nnum)+3     ## negative examples

yyes = np.ones(ynum)
yno = np.ones(nnum)*-1

preX = np.hstack((yes,no))  ## X but no bias yet
y = np.hstack((yyes,yno))   ## targets y

X = np.c_[np.ones(ynum+nnum),preX]  ## adding bias

w = np.linalg.pinv(X).dot(y)
fig = plt.figure(figsize=(9,8))
plt.scatter(X[:ynum,1],np.ones(ynum),c='b',alpha=0.3,s=60)
plt.scatter(X[ynum:,1],-np.ones(nnum),c='r',alpha=0.3,s=60)
plt.plot(X[:,1],X.dot(w),label=r"regression line $\bar{w}=X^\dagger \bar{y}$")
plt.plot(np.linspace(-7,5),np.zeros(50),label="threshold")
mc = np.sign(X.dot(w))!=y
plt.scatter(X[mc,1],np.ones(np.sum(mc)),c='yellow',alpha=1,s=60,label="misclassified")
plt.legend()
plt.title("Some 1D data, regression for classification")
plt.show()
print("{} points misclassified".format(np.sum(mc)))
print("w =",w)
```

```python
## Classification error and regression error

import mystuff as ms
E_in(X,w,y), ms.R2(X.dot(w),y)
```

### Applying SGD

When we did regression above, we used the normal equations to find $\bar{w}$.

$$w = np.linalg.pinv(X).dot(y)$$

Instead of doing that, let's use SGD to find a $\bar{w}$ minimizing the OLS error.

First we make a plot of the OLS error function, below.


```python
xx = np.linspace(-10,10,100)
yy = np.linspace(-10,10,100)

XX,YY = np.meshgrid(xx,yy)
W = np.c_[XX.ravel(),YY.ravel()]
errs = ((X.dot(W.T)).T - y)**2
E = np.mean(errs,axis=1)
plt.contourf(XX,YY,E.reshape(100,100),levels=300)
plt.title(r"Contour plot of $z=E_{in}(\bar{w})$ (regression)")
plt.xlabel(r"$w_0$")
plt.ylabel(r"$w_1$")
plt.colorbar()
plt.show()
```

### Could we do better?

This way of using regression for classification is kind of lazy.

Because the error measures do not quite agree on what "best" means

the "best" regression $\bar{w}$ is not quite the best classification $\bar{w}$.

If we look at the dataset we can see that there exists a $\bar{w}$ with zero classification error.


```python
Xgood = X[:ynum,1]
Xbad = X[ynum:,1]
a = np.array([[w[1]]])
b = np.array([[-w[0]]])
s=np.linalg.solve(a,b)
print("rightmost blue point = ",np.max(Xgood))
print("leftmost red point = ",np.min(Xbad))
print("Actual zero of line corresponding to w = ",s[0][0])
```
