### Logistic regression

Earlier we thresholded linear regression to make a classifier.

First we learned $\bar{w} = X^\dagger\bar{y}$, where $X^\dagger$ is the pseudoinverse.

Then, with the threshold $a=0$, we defined this hypothesis:

$$h_\bar{w}(\bar{x}) = \begin{cases} +1 \,\text{if}\, \bar{w}^T\bar{x} > a \\ 
                                    -1 \,\text{if}\, \bar{w}^T\bar{x} \leq a\end{cases}$$

This is the same as

$$\hat{y} = sign(\bar{w}^T\bar{x}).$$

This sort of abrupt cutoff is sometimes called a **hard threshold**.

We didn't minimize classification error directly because it is too blocky a function.

There are a lot of flat abrupt plateaus and no slopes, and gradient descent can't operate.

But what regression tries to optimize (linear fit) is not exactly the same as what classification seeks to optimize (best decision boundary).

Therefore we didn't find the optimal classification boundary in the previous slides.


---

We now examine a sort of compromise that keeps the best properties of each:  

The error function will be continuous, allowing optimization using calculus (gradient descent).

But the error function also is closer to ordinary classification error.

---

To accomplish this we will use a **soft threshold**.

$$h_\bar{w}(\bar{x}) = \theta(\bar{w}^T\bar{x})$$

where $\theta(s) = \frac{e^s}{1+e^s} = \frac{1}{1+e^{-s}}$.

Rather than abruptly changing from -1 to +1, this hypothesis gently changes from 0 to 1.

By introducing continuous change, it becomes possible to use gradient descent to find the best $\bar{w}$.


```python
import matplotlib.pyplot as plt
import numpy as np

xx = np.linspace(-7,7)
yy = np.exp(xx)/(1+np.exp(xx))

plt.plot(xx,yy)
plt.title("The logistic, or sigmoid function")
plt.show()
```

### Logistic Regression

The basic idea of logistic regression to use $\theta(\bar{w}^T\bar{x})$ as a model for $P(y=+1|\bar{x})$.

That is, we assume that

$$h_\bar{w}(\bar{x}) = \theta(\bar{w}^T\bar{x}) \approx P(y=+1 | \bar{x}).$$

We want to find the $\bar{w}$ that fits this assumption the best.

There is a somewhat crazy strategy for doing this that requires a second of thought.

This approach is called the method of [*maximum likelihood*](https://en.wikipedia.org/wiki/Maximum_likelihood_estimation).

Statement:  Assume that $P(\bar{x})$ and $\theta(\bar{w}^T\bar{x})$ produced the data $(\bar{x}_1,y_1),(\bar{x}_2,y_2),\ldots$, but we don't know for sure what $\bar{w}$ actually was.  

**Which $\bar{w}$ maximizes the chance that we would see the data that we actually do see?**

---

We now construct a measure of how the data is for a given $\bar{w}$.

This will basically be our error function for $\bar{w}$.

The $\bar{w}$ that maximizes the likelihood of seeing the data will be the final hypothesis.

---

Under the assumption that $\theta(\bar{w}^T\bar{x}) \approx P(y=+1 | \bar{x})$,

$$P(y | \bar{x}) = \begin{cases} \theta(\bar{w}^T\bar{x}) \text{ for } y=+1\\ 1-\theta(\bar{w}^T\bar{x}) \text{ for } y=-1 \end{cases}$$

A nice property of the logistic function helps simplify this expression:

$$1-\theta(s) = 1-\frac{e^s}{1+e^{s}} = \frac{1+e^s-e^{-s}}{1+e^{s}} = \frac{1}{1+e^s}$$

Therefore

$$1-\theta(s) = \theta(-s)$$

and

\begin{align}
P(y | \bar{x}) &= \begin{cases} h_\bar{w}(\bar{x}) \text{ for } y=+1\\ 1-h_\bar{w}(\bar{x}) \text{ for } y=-1 \end{cases}\\ &= \begin{cases} \theta(\bar{w}^T\bar{x}) \text{ for } y=+1\\ 1-\theta(\bar{w}^T\bar{x}) \text{ for } y=-1 \end{cases}\\ &= \begin{cases} \theta(\bar{w}^T\bar{x}) \text{ for } y=+1\\ \theta(-\bar{w}^T\bar{x}) \text{ for } y=-1 \end{cases}\\
&= \theta(y\bar{w}^T\bar{x})
\end{align}

The dataset is supposed to be independently generated:

$(\bar{x}_1,y_1),(\bar{x}_2,y_2),\cdots,(\bar{x}_N,y_N)$.

The probability that we see all the $y$'s that we actually do see given the $\bar{x}$'s is therefore just the product:

$$\prod_{n=1}^N P(y_n|\bar{x}_n) \approx \prod_{n=1}^N \theta(y\bar{w}^T\bar{x})$$

We want to maximize the right hand side by selecting the right $\bar{w}$.

But we can equivalently minimize the negative logarithm.  

This works better both numerically and analytically.

\begin{align}
-\frac{1}{N}\ln(\prod_{n=1}^N P(y_n|\bar{x}_n)) &= -\frac{1}{N}\sum_{n=1}^N \ln(P(y_n|\bar{x}_n))\\
&= \frac{1}{N}\sum_{n=1}^N \ln(\frac{1}{P(y_n|\bar{x}_n)})\\
&= \frac{1}{N}\sum_{n=1}^N \ln(\frac{1}{\theta(y\bar{w}^T\bar{x})})\\
&= \frac{1}{N}\sum_{n=1}^N \ln(1+e^{-y_n\bar{w}^T\bar{x}_n})
\end{align}

Notice that this is just the average of the pointwise error function

$${\bf e}(\bar{w}) = \ln(1+e^{-y_n\bar{w}^T\bar{x}_n})$$

We can see that ${\bf e}(\bar{w})$ is big when $\bar{x}_n$ is misclassified and small when $\bar{x}$ is classified correctly.

When $\bar{x}_n$ is miscalssified $y_n$ and $\bar{w}^T\bar{x}$ are different signs.

But when the classification is correct, the signs are the same. 

To summarize, the in-sample error is

$$E_{in}(\bar{w}) = \frac{1}{N}\sum_{n=1}^N \ln(1+e^{-y_n\bar{w}^T\bar{x}_n})$$

and the pointwise error is

$${\bf e}(\bar{w}) = \ln(1+e^{-y_n\bar{w}^T\bar{x}_n})$$

and $E_{in}$ is the average of ${\bf e}(\bar{w})$ over all the training data.


### What's the gradient?

As we said earlier, training in machine learning usually just means finding the solution to the equation

$$\nabla E_{in}(\bar{w}) = \bar{0}.$$

As the book explains, for the logistic regression error function,

$$\nabla E_{in}(\bar{w}) = \frac{1}{N}\sum_{n=1}^N \nabla {\bf e}$$

and

$$\nabla {\bf e} = -\frac{y_n\bar{x}_n}{1+e^{y_n\bar{w}^T\bar{x}}}$$

The zeros of this expression can't be found analytically the way we did for the linear regression gradient.

Instead we use gradient descent with $\nabla E_{in}$ or stochastic gradient descent with $\nabla {\bf e}$.

### Hard thresholding

Logistic regression is called 'regression' because it estimates the continous valued function $P(y=+1 |\bar{x})$.

But it is really a *classification* technique.  

Suppose you give logistic regression a new instance $\bar{x}$ and it gives you back $\hat{p} \approx P(y=+1| \bar{x})$.  

You can then *threshold* this $\hat{p}$ to get a definite decision:

If $\hat{p} \geq \frac{1}{2}$ then predict $y=+1$.

If $\hat{p} <\frac{1}{2}$ then predict $y=-1$.

This is what we did for the images above.

We obtained a perfect classification score on our toy data.

```python
mc = 2*(sigmoid(X.dot(w)) > 0.5)-1 !=y
mc
```

### Other possibilities

Sometimes you might **want** to predict a probability rather than a definite value.

Actually thresholding $\hat{p}$ destroys information.

Here is a plot of some data with a "soft" decision boundary.

In the yellow region we are confident to predict +1.

In the blue region we are confident predicting -1.

In the green region we know we might be wrong.



```python
from boundary import softboundary
plt=softboundary()
plt.show()
```

### Other thresholds

It's natural to threshold $\hat{p}$ at $\frac{1}{2}$ in the absence of other information.

But suppose that $\hat{p}$ is the probability that someone does **not** have cancer given a certain test.

Then you might want to threshold at $\frac{95}{100}$ rather than $\frac{1}{2}$ because one type of error is much worse than the other.

In general adjusting the threshold gives a tradeoff between [precision and recall](https://en.wikipedia.org/wiki/Precision_and_recall).

In a later lesson we might study this tradeoff in more detail.

```python
sevens = (y=="7")
twos = (y=="2")
sevenORtwo = sevens+twos
Xt = X[sevenORtwo]
yt = y[sevenORtwo]
yt = (yt=="7")*2-1

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
#scaler = MinMaxScaler()


Xt = np.hstack((np.ones((Xt.shape[0],1)),Xt))  ## bias column

X_train,X_test,y_train,y_test = train_test_split(Xt,yt)



scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

```

```python
def E_in(X,y,w):
    mc = (np.sign(X.dot(w)) != y)
    return np.sum(mc)/len(mc)

def Pocket(X,y,w,max_iter=1000):
    E_in_best = 1
    E_in_argbest = w
    for i in range(max_iter):
        mc = (np.sign(X.dot(w)) != y)
        if not mc.any():
            break
        if E_in(X,y,w) < E_in_best:
            E_in_best = E_in(X,y,w)
            E_in_argbest = w
            
        badx = X[mc][0]
        bady = y[mc][0]
        w = w+bady*badx
    return E_in_argbest,i

w = np.zeros(X_train.shape[1])

w,i=Pocket(X_train,y_train,w)

print("Finished in {} iterations".format(i))
print("E_in = {}".format(E_in(X_test,y_test,w)))
```
