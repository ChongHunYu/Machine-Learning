```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1)[1:-1]
f1 = -np.log(x)
f2 = -np.log(1-x)
plt.plot(x,f1,label=r"if $y=+1$")
plt.plot(x,f2,label=r"if $y=-1$")
plt.xlabel(r"$\hat{p}$")
plt.ylabel(r"Error")
plt.legend()
plt.show()
```

```python
x = np.linspace(-7,7)
plt.plot(x,1/(1+np.exp(-x)))
plt.title(r"Logit, $\theta(s)=\frac{1}{1+e^{-s}}$")
plt.show()
```

```python
x = np.linspace(-7,7)
plt.plot(x,np.arctan(x))
plt.title(r"Arctangent")
plt.show()
```

```python
x = np.linspace(-7,7)
plt.plot(x,np.tanh(x))
plt.title(r"Hyperbolic Tangent")
plt.show()
```
