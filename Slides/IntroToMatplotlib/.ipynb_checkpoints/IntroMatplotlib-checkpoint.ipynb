### Introducing matplotlib

The matplotlib library is commonly used for visualizations in ML.

There are other rival plotting packages (eg Bokeh and Seaborn), but you really need to understand matplotlib to use them.

The "basic idea" of the way you plot in matplotlib is common across a wide range of plotting libraries.  

Here is an informative discussion of the main differences between matplotlib, seaborn and bokeh:

https://www.reddit.com/r/Python/comments/4tuwoz/how_do_you_decide_between_the_plotting_libraries/


### The way you give data to matplotlib

Most plots are two dimensional, and therefore consist of an $x$ and a $y$ axis.

The things that get plotted are points which are pairs $(x,y)$ of particular values.

In matplotlib you give the plotting program two vectors of data of the same length: 

* An "x" vector $\bar{x}$
* A "y" vector $\bar{y}$

The $x_i$ entry of $\bar{x}$ corresponds to the $y_i$ entry of $\bar{y}$ to give the data point $(x_i,y_i)$.



### Example

Below I give an example of how this works in practice.



```python
x = [1,2,3]
y = [1,5,-2]

print("The data points are:")
for xx,yy in zip(x,y):
    print("({},{})".format(xx,yy))
```

### Mumbo jumbo

As with any software library, there is some arbitrary syntactical stuff that you just need to get used to.

For matplotlib this has to do with the "pyplot" object, which we call **plt** by convention.

The program below shows a minimal line plot made with matplotlib.


```python
import matplotlib.pyplot as plt

plt.plot(x,y)

plt.show()
```

### A line plot? Or scatterplot?

Notice that matplotlib assumed that the data points were just a sample from a smooth function and tried to "connect the dots" with a line.

Often you just want to plot the points themselves.

This is called a **scatterplot** and it works as follows.


```python
import matplotlib.pyplot as plt

plt.scatter(x,y)

plt.show()
```

### Bells and whistles

Below I will gussy up our basic scatterplot to demo some further features of matplotlib.



```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8, 8))                      

plt.title("The secret of the universe")
plt.scatter(x,y,color=['b','r','g'],s=[40,80,1600],alpha=0.5)
plt.xlabel("Ethanol per juicebox")
plt.ylabel("Misery")
plt.xticks([1,2,3,4,5], ('Tom', 'Dick', 'Harry', 'Sally', 'Sue'),rotation=90)
s = ["huey","dewey","louie"]
for xx,yy in zip(x,y):
    plt.text(xx+0.2,yy-0.2,s[-1],fontsize=20,rotation=-15)
    s.pop()

plt.show()
```

###  Scatterplots with labels

In ML we will look at lots of scatterplots.  

Often different points are of different types.

Below we steal a nice example of that from [here](https://stackoverflow.com/questions/17411940/matplotlib-scatter-plot-legend)



```python
import matplotlib.pyplot as plt
from numpy.random import random

colors = ['b', 'c', 'y', 'm', 'r']

fig = plt.figure(figsize=(10, 10))                      


lo = plt.scatter(random(10), random(10), marker='x', color=colors[0])
ll = plt.scatter(random(10), random(10), marker='o', color=colors[0])
l  = plt.scatter(random(10), random(10), marker='o', color=colors[1])
a  = plt.scatter(random(10), random(10), marker='o', color=colors[2])
h  = plt.scatter(random(10), random(10), marker='o', color=colors[3])
hh = plt.scatter(random(10), random(10), marker='o', color=colors[4])
ho = plt.scatter(random(10), random(10), marker='x', color=colors[4])

plt.legend((lo, ll, l, a, h, hh, ho),
           ('Low Outlier', 'LoLo', 'Lo', 'Average', 'Hi', 'HiHi', 'High Outlier'),
           scatterpoints=1,
           loc='lower left',
           ncol=3,
           fontsize=8)

plt.show()
```

### Plotting smooth functions

Often, especially in regression, we might want to plot a smooth curve.  

Below we illustrate how to do this using numpy universal functions.  

A **universal function** is a function that is applied to a vector of values in an element-by-element fashion.



```python
import numpy as np

A = np.array([1,2,3,4])
A
```

```python
## The sine function will be applied to each vector element "respectively"
np.sin(A)
```

```python
## The arctan function will be applied to each vector element "respectively"

np.arctan(A)
```

```python
## The sigmoid function will be applied to each vector element "respectively"


sig = 1/(1+np.exp(-A))
sig
```

```python
## A very rough plot of the above functions
fig = plt.figure(figsize=(8, 8))                      

plt.plot(A,np.sin(A),label="sine")
plt.plot(A,np.arctan(A),label="atan")
plt.plot(A,sig,label="sigmoid")
plt.legend()
plt.show()
```

###  Smoothing things out

The plot we just made is very rough because the domain only contains four datapoints.

We can make things smoother by enriching the domain, filling in the gaps.

Below we introduce the usual way of making a smooth interval of values in Python.

```python
### Very rarely you might want to use a logarithmic scale.
### in this case, use logspace.

np.logspace?
```

```python
np.logspace(0,6,num=15)
```

```python
## A very rough plot of the above functions

A = np.linspace(-6,6,1000)
fig = plt.figure(figsize=(8, 8))                      

plt.plot(A,np.sin(A),label="sine")
plt.plot(A,np.arctan(A),label="atan")
sig = 1/(1+np.exp(-A))

plt.plot(A,sig,label="sigmoid")
plt.legend()
#plt.axhline(0, color='black')
#plt.axvline(0, color='black')

plt.show()
```

### Plotting a covariance matrix

As we mentioned in the pandas tutorial, the correlations between different features are very interesting in ML.

It is possible to use matplotlib to create nice visualizations of the correlations (covariance) between the features in a dataset.

Here is an example from [here](https://stackoverflow.com/questions/29432629/plot-correlation-matrix-using-pandas).



```python
import pandas as pd
import numpy as np

rs = np.random.RandomState(0)
df = pd.DataFrame(rs.rand(10, 10))
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
# 'RdBu_r' & 'BrBG' are other good diverging colormaps
```

###  Plotting higher dimensional data

Consider a function of two variables $f(x,y)$.

The graph of such a function is given by the surface $z = f(x,y)$.

Each point in the $x$$y$ plane is associated with a height $z$.

There are two natural ways to plot such data:

* As a contour plot (smooth f)
* A plot with 3 axes

First we show how to create a contour plot.


```python
## This works well when f is smooth 

xvec = np.linspace(-10.,10.,100)                               
X,Y = np.meshgrid(xvec, xvec)                                
fig = plt.figure(figsize=(6, 4))     
Z = X**2 - Y**2
plt.contourf(X, Y, Z,5)                             
plt.colorbar()    
plt.show()
```

### Examples of 3 axis plots

Below we borrow some nice examples from here:

https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

```python
'''
======================
3D surface (color map)
======================

Demonstrates plotting a 3D surface colored with the coolwarm color map.
The surface is made opaque by using antialiased=False.

Also demonstrates using the LinearLocator and custom formatting for the
z axis tick labels.
'''

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```

```python
'''
==============
3D scatterplot
==============

Demonstration of a basic scatterplot in 3D.
'''

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np


def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
'''
===========================
More triangular 3D surfaces
===========================

Two additional examples of plotting surfaces with triangular mesh.

The first demonstrates use of plot_trisurf's triangles argument, and the
second sets a Triangulation object's mask and passes the object directly
to plot_trisurf.
'''

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.tri as mtri


fig = plt.figure(figsize=3*plt.figaspect(0.5))

#============
# First plot
#============

# Make a mesh in the space of parameterisation variables u and v
u = np.linspace(0, 2.0 * np.pi, endpoint=True, num=50)
v = np.linspace(-0.5, 0.5, endpoint=True, num=10)
u, v = np.meshgrid(u, v)
u, v = u.flatten(), v.flatten()

# This is the Mobius mapping, taking a u, v pair and returning an x, y, z
# triple
x = (1 + 0.5 * v * np.cos(u / 2.0)) * np.cos(u)
y = (1 + 0.5 * v * np.cos(u / 2.0)) * np.sin(u)
z = 0.5 * v * np.sin(u / 2.0)

# Triangulate parameter space to determine the triangles
tri = mtri.Triangulation(u, v)

# Plot the surface.  The triangles in parameter space determine which x, y, z
# points are connected by an edge.
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap=plt.cm.Spectral)
ax.set_zlim(-1, 1)


#============
# Second plot
#============

# Make parameter spaces radii and angles.
n_angles = 36
n_radii = 8
min_radius = 0.25
radii = np.linspace(min_radius, 0.95, n_radii)

angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)
angles = np.repeat(angles[..., np.newaxis], n_radii, axis=1)
angles[:, 1::2] += np.pi/n_angles

# Map radius, angle pairs to x, y, z points.
x = (radii*np.cos(angles)).flatten()
y = (radii*np.sin(angles)).flatten()
z = (np.cos(radii)*np.cos(angles*3.0)).flatten()

# Create the Triangulation; no triangles so Delaunay triangulation created.
triang = mtri.Triangulation(x, y)

# Mask off unwanted triangles.
xmid = x[triang.triangles].mean(axis=1)
ymid = y[triang.triangles].mean(axis=1)
mask = np.where(xmid**2 + ymid**2 < min_radius**2, 1, 0)
triang.set_mask(mask)

# Plot the surface.
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_trisurf(triang, z, cmap=plt.cm.CMRmap)


plt.show()
```
